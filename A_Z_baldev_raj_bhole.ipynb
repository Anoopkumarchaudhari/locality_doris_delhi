```python
from bs4 import BeautifulSoup
```

```python
with open('c_baldev_raj_bhola_2.html','r',encoding='utf-8') as f:
    html = f.read()
```

```python
soup = BeautifulSoup(html,'html.parser')
```

```python
rows = soup.find_all("tr")
all_data = []
for row in rows:
    try:
        record = {
            "Reg.No": row.find(id=lambda x: x and "regno" in x).text.strip(),
            "Reg.Date": row.find(id=lambda x: x and "regdate" in x).text.strip(),
            "First Party": row.find(id=lambda x: x and "firstprt" in x).text.strip(),
            "Second Party": row.find(id=lambda x: x and "secondprt" in x).text.strip(),
            "Property Address": row.find(id=lambda x: x and "proadd" in x).text.strip(),
            "Area": row.find(id=lambda x: x and "proarea" in x).text.strip(),
            "Deed Type": row.find(id=lambda x: x and "deedtp" in x).text.strip(),
            "Property Type": row.find(id=lambda x: x and "protp" in x).text.strip(),
        }
        all_data.append(record)
    except:
        continue  # skip rows without data

# Print all records
for item in all_data:
    print(item)
```

```python
import pandas as pd
df = pd.DataFrame(all_data)
df.to_csv("a_baldev_raj_bhola_2.csv", index=False)
```

```python
import pandas as pd
def extract_data(alpha):
    try:
        with open(f'{alpha}_baldev_raj_bhola_2.html','r',encoding='utf-8') as f:
                 html = f.read()
        soup = BeautifulSoup(html,'html.parser')
        rows = soup.find_all("tr")
        rows = soup.find_all("tr")
        all_data = []
        for row in rows:
            try:
                record = {
                    "Reg.No": row.find(id=lambda x: x and "regno" in x).text.strip(),
                    "Reg.Date": row.find(id=lambda x: x and "regdate" in x).text.strip(),
                    "First Party": row.find(id=lambda x: x and "firstprt" in x).text.strip(),
                    "Second Party": row.find(id=lambda x: x and "secondprt" in x).text.strip(),
                    "Property Address": row.find(id=lambda x: x and "proadd" in x).text.strip(),
                    "Area": row.find(id=lambda x: x and "proarea" in x).text.strip(),
                    "Deed Type": row.find(id=lambda x: x and "deedtp" in x).text.strip(),
                    "Property Type": row.find(id=lambda x: x and "protp" in x).text.strip(),
                }
                all_data.append(record)
            except:
                continue  # skip rows without data
        
        # Print all records
        print(f'============================= records for -- {alpha}-- ============================')
        for item in all_data:
            print(item)

        df = pd.DataFrame(all_data)
        df.to_csv(f"{alpha}_baldev_raj_bhola_2.csv", index=False)
        
    except Exception as e:
        print('file not found')
        print(e)
                

                
```
